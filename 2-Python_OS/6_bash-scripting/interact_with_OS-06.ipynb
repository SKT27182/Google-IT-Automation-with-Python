#3-Advance Bash Scripting

while loop

```
#!/bin/bash

n=1

while [ $n -le 5 ]; do

	echo "Iteration number $n"

	((n+=1))

done
```





```
#!/bin/bash

n=0
command = $1   #takes input from command line
while ! $command && [ $n -le 5]; do
	sleep $n
	((n=n+1))
	echo "Retry #$n"
done
```



for loop
